# 子页信息

```python
def getTitle(link):
    title = {}
    res = requests.get(link)
    res.encoding = 'utf-8'
    soup = BeautifulSoup(res.text,'html.parser')
    try:

        title['标题'] = (soup.select('.main-title')[0].text)
        title['时间'] =(soup.select('.date')[0].text)
        title['来源'] =(soup.select('.source')[0].text)
        try:
            title['关键字'] = soup.select('.keywords')[0]['data-wbkey']
        except:
            pass
        article = []
        for p in soup.select('#article  p')[1:-1]:
            article.append(p.text.strip())
        '\n'.join(article)
        title['内容'] = article
        try:
            title['责任编辑'] = soup.select('.show_author')[0].text.split('：')[1]
        except:
            pass
        return title
except:
        print("啥啊？？")
```

# 汇总

```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://news.sina.com.cn/china/')
res.encoding = 'utf-8'
soup = BeautifulSoup(res.text,'html.parser')  
title = soup.select('.left-content-1.marBot a')
list = []
for n in range(len(title))[:200]:
    if len(title[n]['href']) == 63:
        print(title[n].text)
        list.append(getTitle(title[n]['href']))
```

# 数据处理

```python
import pandas as pd
df = pd.DataFrame(list)
```

```python
df.head()
```

## 处理时间格式


```python
df['时间'] = pd.to_datetime(df['时间'],format = '%Y年%m月%d日 %H:%M')
```

###  拿到时间的 年月日

```python
df['时间'].map(lambda e: e.year)
```

###  调整栏位顺序

```python
df = df[['来源','标题','内容','关键字','责任编辑','时间']]
```
