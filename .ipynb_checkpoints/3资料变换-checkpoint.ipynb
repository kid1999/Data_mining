## 新建一个均价属性

```python
df['均价'] = df['总价']*10000 / df['建筑面积']
df.head()
```

## 移除物业费中的后缀

```python
s = '1.5元/平米・月'
s.split('元')[0]
```

```python
# df['物 业 费']
def removeDollar(e):
    return e.split('元')[0]
```

### 使用map函数 处理

对一个list 进行操作

```python
df['物 业 费'].map(removeDollar)
```

使用 lambda 匿名函数 解决


```python
df['物 业 费'].map(lambda e: e.split('元')[0])
```

### 使用apply函数处理

可以选择 对Dataframe的某一行、某一列 操作


```python
da = pd.DataFrame([[60,70,50],[80,79,68],[63,66,82]], columns = ['First','Second','Third'])
da
```

```python
da.apply(lambda e: e.max() - e.min())  #默认axis =0 对列操作
```

```python
?da.apply
```

```python
da.apply(lambda e: e.max() - e.min(),axis = 1)  #axis= 1 对行操作
```

### applymap 方法

对DataFrame的所有元素进行操作

```python
df.applymap(lambda e: np.nan if e == '暂无资料' else e)
```
