```python
import pandas_datareader
df = pandas_datareader.DataReader('BABA',data_source='yahoo',start = '2014-10-01')
```

```python
df.head()
```

### 使用jupyter 自带的图形化应用

```python
%pylab inline
```

```python
df.plot
```

```python
df['Close'].plot(kind='line',figsize=[10,5],legend=True,title='BABA',grid =True)
```

### 绘制移动平均线

```python
# df['mvg30']
df['mvg30'] = df['Close'].rolling(window=30).mean()
```

```python
df.head()
```

```python
df[['Close','mvg30']].plot(kind ='line',legend=True, figsize=[10,5])
```

### 绘制交易量直方图

```python
df.loc[df.index >= '2018-10-01', 'Volume'].plot(kind = 'bar',title = 'BABA',legend = 'volume')
```
